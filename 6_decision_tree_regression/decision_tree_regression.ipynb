```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
print(data)
```

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
print(x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor

reg = DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
y_pred = reg.predict([[6.5]])
print(y_pred)
```

```python
x_points = np.arange(min(x),max(x),0.1)
print(x_points)
x_points = x_points.reshape(len(x_points),1)
print(x_points)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x_points,reg.predict(x_points),color='blue')
plt.title('Position Salaries Predictions')
plt.xlabel('Position') 
plt.ylabel('Salaries')
```
