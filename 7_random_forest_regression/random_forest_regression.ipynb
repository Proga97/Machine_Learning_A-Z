```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
print(data)
```

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
print(x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()
```

```python
reg.fit(x,y)
```

```python
y_pred = reg.predict([[6.5]])
print(y_pred)
```

```python
x_smooth = np.arange(min(x),max(x),0.1)
print(x_smooth)
x_smooth = x_smooth.reshape(-1,1)
print(x_smooth)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x_smooth,reg.predict(x_smooth),color='red')
plt.title('Position Salaries Predictions')
plt.xlabel('Position') 
plt.ylabel('Salaries')
```
