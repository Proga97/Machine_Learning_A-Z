```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
print(x,y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.25)
print(x_train,x_test,y_train,y_test)
```

```python
from sklearn.preprocessing import StandardScaler

se = StandardScaler()
x_train = se.fit_transform(x_train)
x_test = se.transform(x_test)
print(x_train,x_test)
```

```python
from sklearn.svm import SVC

classifier = SVC()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(y_pred)
y_merge = np.concatenate((y_pred.reshape(-1,1),y_test.reshape(-1,1)),1)
print(y_merge)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

matrix = confusion_matrix(y_true= y_test, y_pred=y_pred)
score = accuracy_score(y_pred=y_pred,y_true=y_test)
print(matrix,score)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = se.inverse_transform(x_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(se.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
