```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
print(data)
```

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
print(x,y)
```

```python
l_reg = LinearRegression()
l_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

p_reg = PolynomialFeatures(degree=2)
x_poly = p_reg.fit_transform(x)
print(x_poly)
```

```python
l_p_reg = LinearRegression()
l_p_reg.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,l_reg.predict(x),color='red')
plt.title('Linear Reg')
plt.xlabel('Position') 
plt.ylabel('Salary') 
plt.show()
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,l_p_reg.predict(x_poly),color='red')
plt.title('Ploy Reg')
plt.xlabel('Position') 
plt.ylabel('Salary') 
plt.show()
```

```python
p_reg_2 = PolynomialFeatures(degree=3)
x_poly_2 = p_reg_2.fit_transform(x)
l_p_reg_2 = LinearRegression()
l_p_reg_2.fit(x_poly_2,y)
plt.scatter(x,y,color='blue')
plt.plot(x,l_p_reg_2.predict(x_poly_2),color='red')
plt.title('Ploy Reg')
plt.xlabel('Position') 
plt.ylabel('Salary') 
plt.show()
```

```python
p_reg_3 = PolynomialFeatures(degree=4)
x_poly_3 = p_reg_3.fit_transform(x)
l_p_reg_3 = LinearRegression()
l_p_reg_3.fit(x_poly_3,y)
plt.scatter(x,y,color='blue')
plt.plot(x,l_p_reg_3.predict(x_poly_3),color='red')
plt.title('Ploy Reg')
plt.xlabel('Position') 
plt.ylabel('Salary') 
plt.show()
```

```python
s_1 = l_reg.predict([[6.5]])
s_2 = l_p_reg.predict(p_reg.fit_transform([[6.5]]))
s_3 = l_p_reg_2.predict(p_reg_2.fit_transform([[6.5]]))
s_4 = l_p_reg_3.predict(p_reg_3.fit_transform([[6.5]]))
print(s_1,s_2,s_3,s_4)
```
